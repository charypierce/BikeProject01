```python
#Convieres a formato fecha para poder trabajar con operaciones
df['started_at']=pd.to_datetime(df.started_at)#cast str to Timestamp
df['ended_at']=pd.to_datetime(df.ended_at)
```

```python
#Dividimos en fecha y hora
df['date']=pd.to_datetime([d.date() for d in df.started_at])#dividir en fecha y convertir a fecha
df['month']=[d.month for d in df.started_at]
df['hours']=[d.time() for d in df.started_at]#Dividir en horas 
df['hour']=[d.hour for d in df.started_at]#hora 
```

```python
#Vemos cuantos datos son inconsistentes en fechas, en este caso hay 139 datos que tienen un mal formato en la fecha
print(df[df.started_at>df.ended_at].date.value_counts().sum())
data1=df[df.started_at>df.ended_at].date.value_counts()
data1[:10]
#Al ver el recuento de fechas con error, nos damos cuenta que el 38% de errores ocurre el día 7 de noviembre del 2021
#Una busqueda rápida por internet nos muestra que el horario de verano Termina el día 7 de noviembre del 2021 "2021-11-07"
#Inicia el día 14 de marzo del 2021  2021-03-14#Por lo q probablemente se registe una hora más en los ciclistas
#Pero no esta en el rango análisado la fecha 2021-03-14
#el domingo 13 de marzo de 2022, a las 2:00 a.m. empieza el horario de verano nuevamente
#el domingo 6 de noviembre de 2022, inicia horario de invierno#No esta en rango análisado
```

```python
df.loc[(df['date']=='2021-11-07')&(df['started_at']>df['ended_at'])].head(1)
```

```python
#Para corregir esta fecha restaremos 1 hora a started_at para la fecha '2021-11-07' y donde la
#hora de inicio sea mayor a la del final.
index1=df.loc[(df['date']=='2021-11-07')&(df['started_at']>df['ended_at'])].index
```

```python
df.loc[index1,'started_at']=df.loc[index1].started_at-pd.to_timedelta(1, unit='h')
```

```python
df.loc[index1].head(1)
```

Dádo que cambiamos la fecha al restar 1 hora al horario, la columna de horas se vio afectada por lo que volveremos a generarlas, tambien generaremos las fechas por si la hora era cerca de media noche entonces cambiará el día.
Como no aumentamos ni meses, o años las demas columnas permanecen igual.
Generaremos 

```python
df.loc[index1,['date']] = pd.to_datetime([d.date() for d in df.loc[index1].started_at])
df.loc[index1,['hours']]= [d.time() for d in df.loc[index1].started_at]#Dividir en horas
df.loc[index1,['hour']]=[d.hour for d in df.loc[index1].started_at]#hora
```

```python
#Aqui Empieza el horario de verano por lo que a las 2am se adelanta el reloj 1 hra. por lo que
#aquelos que empezaron antes de las 2 y terminaron despues de las 3, tienen agregado la hora de horario de verano
#Esto sesga los datos por lo que deben ser corregidos
index2=df.loc[(df['date']=='2022-03-13')&(df.started_at<='2022-03-13 02:00:00') & (df.ended_at>='2022-03-13 03:00:00')].index
```

```python
df.loc[index2].head(1)
```

```python
#Lo arreglaremos sumando 1hr a 'started_at'
#Abajo esta más sintetizado 
df.loc[index2,'started_at']=df.loc[(df['date']=='2022-03-13')&(df.started_at<='2022-03-13 02:00:00') & (df.ended_at>='2022-03-13 03:00:00'),['started_at']]+pd.to_timedelta(1, unit='h')
```
